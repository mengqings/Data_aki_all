```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
data=pd.read_csv("F:\TrainTest\AKI\Data\data1.csv")
data
```

```python
#实现归一化
scaler = MinMaxScaler() #实例化
scaler = scaler.fit(data) #fit，在这里本质是生成min(x)和max(x)
result = scaler.transform(data) #通过接口导出结果
result
```

当数据(x)按均值(μ)中心化后，再按标准差(σ)缩放，数据就会服从为均值为0，方差为1的正态分布（即标准正态分
布），而这个过程，就叫做数据标准化(Standardization，又称Z-score normalization)，

数据的标准化

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
data1=pd.read_csv("F:\TrainTest\AKI\Data\data1.csv")
data1
```

```python
#标准化
scaler=StandardScaler()
#实例化
scaler.fit(data1)
#查看均值属性
scaler.mean_
```

```python
#查看方差属性
scaler.var_
```

```python
#接口导出结果
x_Std=scaler.transform(data1)
```

```python
#导出的结果是一个数组，用mean()查看均值
x_Std.mean()
```

```python
#用std()查看方差
x_Std.std()
```
